```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
#to show our output inside the jupyter notebook
#without this command results sometimes show up in pop up windows
```

```python
df = pd.read_csv("vgsales.csv")
df.head()
```

```python
#to add a style of grid in the background
sns.set_style("whitegrid")
```

```python
#Drawing a line chart
#to plot multiple lines in the same graph call the function plt.plot multiple times with the d/t data in the same cell
plt.plot(df["Year"],df["Global_Sales"],marker="x",color="green");
plt.plot(df["Year"],df["Other_Sales"],marker="o",color="red")
plt.xlabel("year")
plt.ylabel("Global_Sales")
plt.legend(["Global-sales","Other-sales"])
plt.show()
```

```python
#to add a dark background grid
sns.set_style("darkgrid")
```

```python
plt.plot(df["EU_Sales"],df["Year"])
plt.xlabel("Eu_sales")
plt.ylabel("year")
plt.title("EU_sales in each year")
plt.legend(["EU_sales"])
plt.show()
```
